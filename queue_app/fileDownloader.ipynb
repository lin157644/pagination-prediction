```python
import sys
import os
```

```python
sys.path.insert(0, os.path.abspath(r".."))
from autopager.preprocessing import generate_page_component
```

```python
file = pd.read_csv('./needCrawl.csv')
```

```python
file = file[['File Name','Page URL']]
```

```python
for idx, row in file.iterrows():
    file_idx = row['File Name']
    page_url = row['Page URL']
    print(f"Now downloading page {file_idx}: {page_url}")
    try:
        page_html = generate_page_component(page_url)['html']
        with open(f"./download/{str(file_idx)}.html", 'w') as f:
            f.write(page_html)
    except Exception as e:
        print(f"page idx: {file_idx} have error ({e})")
    print(f"Finish saving page {file_idx}: {page_url}")
```
